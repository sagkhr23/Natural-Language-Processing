```python
import pandas as pd
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
```

```python
data_train = pd.read_table('train-data.csv')
data_test = pd.read_table('test-data.csv')
```

```python
def process(data):
  data.columns = ['Sno','Class','Text','Random']
  cats = data['Class'].apply(lambda x: x.split(':')[0])
  sub_cats = data['Class'].apply(lambda x: x.split(':')[1])
  Text = data['Text'].tolist()
  count_vect = CountVectorizer()
  newText = []
  stops = set(stopwords.words('english'))
  stop=[]
  for p in Text:
      s = [d.lower() for d in p.split()]
      newText.append(' '.join([i for i in s if i not in stop]))

  X = count_vect.fit_transform(newText)
  Y=[]
  for i in range(0,len(pd.get_dummies(cats))):
      Y.append(pd.get_dummies(cats).iloc[i].tolist())
  X=X.toarray()
  return X,Y

X_test,Y_test = process(data_test)
X_train,Y_train = process(data_train)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers.core import Dropout
classifier = Sequential()
classifier.add(Dense(output_dim = 1000, init = 'uniform', activation = 'sigmoid', input_dim = 8409))
classifier.add(Dropout(0.2))
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'sigmoid'))
classifier.add(Dropout(0.2))
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
classifier.fit(X_train, Y_train, batch_size = 10, nb_epoch = 1,validation_split=.05)
```
